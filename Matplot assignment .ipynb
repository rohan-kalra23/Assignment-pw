# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x=np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
```

```python
plt.scatter(x,y , c='c')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("this is a scatter plot ")
```

what is scatter plot -->>  scatter plot is the grpahical representation of two dataset (where both the dataset are having numerical value ) , like in the above graph we can see it take the x value on the x axis and the y value are on the y axis and y values are being shown here with respect to the values of x . so we can say whenever we have dataset in the numerical value and we want to see the insights of both the data with repect to each other we can use the  scatter plot 

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
x=np.array([0,1,2,3,4,5])
y = np.array([0,100,200,300,400,500])
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([50,20,40,20,60,70])
x2=np.array([0,1,2,3,4,5])
y2 =np.array([10,20,30,40,50,60])
x3=np.array([0,1,2,3,4,5])
y3 = np.array([200,350,250,550,450,150])
```

```python
plt.plot(x,y , c='c')
plt.plot(x1,y2 , c='r')
plt.plot(x2,x3 , c='g')
plt.plot(x3 , y3 , c='b')
```

why is subplot function used --- >> in the above graph we can see when we try to visualize 4 graphs at the same time then all the four graphs are over lapping each other here and to avoid this thing and to make 4 graphs differently in the same ouput subplot() function is used in the matplot library .

```python
# now use the subplot function 
plt.plot(x,y , c='c')
plt.subplot(2,2,1)

plt.plot(x1,y1 , c = 'r')
plt.subplot(2,2,2)

plt.plot(x2,y2, c= 'g')
plt.subplot(2,2,3)

plt.plot(x3, y3 , c='b')
plt.subplot(2,2,4)
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Bar plot -->> A bar plot, also known as a bar chart or bar graph, is a type of chart that presents categorical data with rectangular bars. The length of each bar represents the value or frequency of a particular category. Bar plots are commonly used to compare and display discrete data sets or to visualize the distribution of a categorical variable

why it is used -->> a bar plot is used to look at many factors of the data in an effective way like we can do the comparison b/w two data eadily and we can also get to know the trend over time like what is the trend changing , these type of changes can be observed easily using the bar graph .


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company , profit , color = 'c' )
plt.title("bar plot")
plt.xlabel("company name")
plt.ylabel("profit this year ")
plt.show()
```

```python
plt.barh(company , profit , color = 'c' )
plt.title("bar plot")
plt.xlabel("Profit")
plt.ylabel("company name ")
plt.show()
```
